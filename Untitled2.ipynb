```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

cap = cv2.VideoCapture('data/train.mp4')
file_speed = open('data/train.txt', 'r') 

x=100
y=100
h=100
w=100

nc = 5
samples = []
speeds = []
i=0
sample = np.zeros((h, w, nc))
speed = np.zeros(nc)

while(cap.isOpened()):
    ret, frame = cap.read()
    line = file_speed.readline()
    
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    crop_img = gray[y:y+h, x:x+w].copy()
    
    cv2.imshow('frame',crop_img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
    sample[:, :, i] = crop_img
    speed[i] = float(line)
    
    i += 1

    if i%nc==0:
        i = 0
        samples.append((sample).copy())   
        speeds.append(np.mean(speed))
    

cap.release()
cv2.destroyAllWindows()
```

```python
def correl_mismatch(slice0, slice1):
    """ Negative correlation between the two images, flattened to 1D """
#     m = -np.corrcoef(slice0.ravel(), slice1.ravel())[0, 1]
    m = np.mean(np.abs(slice0-slice1))
    return m

index = 32
img1 = samples[index][:,:,0]
img2 = samples[index][:,:,1]
correl_mismatch(img1, img2)
```

```python
def fancy_x_trans_slice(img_slice, x_y_trans):
    # Resample image using bilinear interpolation (order=1)
    trans_slice = snd.affine_transform(img_slice, [1, 1], -x_y_trans, order=3)
    return trans_slice

def cost_fcn(t, img1, img2):
    shifted = fancy_x_trans_slice(img2, t)
    mismatch = correl_mismatch(img1, shifted)
    return mismatch
```

```python
def get_test_train(samples, speeds, train_prc):
    n_train = int(len(samples)*.9)
    samples_array = np.asarray(samples)
    x_train = samples_array[:n_train]
    x_test = samples_array[n_train:]
    y_train = np.asarray(speeds[:n_train])
    y_test = np.asarray(speeds[n_train:])
    return x_train, y_train, x_test, y_test
```

```python
from tensorflow import keras as keras
# from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.layers import Conv3D, MaxPooling3D, Conv2D, MaxPooling2D, Reshape
from tensorflow.keras import backend as K

batch_size = 128
epochs = 12

# input image dimensions
img_rows, img_cols, channels = 100, 100, 5

# the data, split between train and test sets
# (x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, y_train, x_test, y_test = get_test_train(samples, speeds, 0.8)
input_shape = (img_rows, img_cols, channels, 1)


x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
x_train = x_train.reshape(x_train.shape[0], img_rows, img_cols, channels, 1)
x_test = x_test.reshape(x_test.shape[0], img_rows, img_cols, channels, 1)

print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')


# convert class vectors to binary class matrices

model = Sequential()
model.add(Conv3D(32, kernel_size=(3, 3, channels), activation='relu', input_shape=input_shape))
# model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=input_shape))
model.add(Reshape((98, 98, 32), input_shape=(98, 98, 1, 32)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1))

model.compile(loss=keras.losses.mean_squared_error,
              metrics=['MeanSquaredError'])

model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(x_test, y_test))
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
